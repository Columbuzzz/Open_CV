```python
import os
import cv2
```

First method .... There are 4methods 

```python
img = cv2.imread(os.path.join('.','data','dog.99.jpg'))
k_size = 7
img_blur = cv2.blur(img,(k_size, k_size))
cv2.imshow('img', img)
cv2.imshow('blur_img', img_blur)
cv2.waitKey(0)
```

```python
Gaussian_blur = cv2.GaussianBlur(img , (k_size,k_size) , 3)
cv2.imshow('gaussian_blur', Gaussian_blur)
```

```python
median_blur = cv2.medianBlur(img,k_size)
cv2.imshow('median_blur', median_blur)
```

```python
# Removing noise
path = os.path.join('.', 'data', 'image_noise.jpg')
img = cv2.imread(path)
k_size = 3
cv2.imshow('img', img)
cv2.waitKey(400)

```

```python
img_blur = cv2.medianBlur(img , k_size)
cv2.imshow('blur_img', img_blur)
cv2.waitKey(0)
```
